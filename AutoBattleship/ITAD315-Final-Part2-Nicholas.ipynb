# Project Description

My project is to implement a simple version of battleship where random spots are chosen for the ships, and it uses discrete probability and statistical analysis to determine its next move.  There will be a probability grid representing the likelihood that each cell on the board contains part of an opponent’s ship. The AI will select its next shot using weighted random selection. The cells that are more likely to contain a ship based on the probability grid are more likely to be selected next. The project will demonstrate how discrete mathematics and stats can be applied to games.

```python
import numpy as np
import random
import matplotlib.pyplot as plt

GRID_SIZE = 10
# Carrier, Battleship, Cruiser, Submarine, Destroyer
SHIP_SIZES = [5, 4, 3, 3, 2] 

# the first thing we would do is to place the ships randomly on the board and check if the placement is valid
def place_ships():
    board = np.zeros((GRID_SIZE, GRID_SIZE), dtype=int)

    def space_is_free(row, col):
        return (
            0 <= row < GRID_SIZE
            and 0 <= col < GRID_SIZE
            and board[row, col] == 0
        )

    # Check the 3x3 area around the cell to enforce the 'no adjacent ships' rule
    def area_is_clear(row, col):
    # Check every surrounding cell in a 3×3 block centered on (row, col)
        for row_offset in [-1, 0, 1]:       # change in row: up, same, down
            for col_offset in [-1, 0, 1]:   # change in col: left, same, right

                check_row = row + row_offset
                check_col = col + col_offset

                # Ignore out-of-bounds checks (edges of the grid)
                if 0 <= check_row < GRID_SIZE and 0 <= check_col < GRID_SIZE:

                    # If ANY surrounding cell already contains a ship (1), it's not clear
                    if board[check_row, check_col] == 1:
                        return False

    # If we finished all checks, the surrounding area is empty
    return True


    def can_place(start_row, start_col, size, direction):
        for i in range(size):
            ship_row = start_row + (i if direction == "V" else 0)
            ship_col = start_col + (i if direction == "H" else 0)

            # Check if the ship segment is in bounds, free, and clear of other ships
            if not space_is_free(ship_row, ship_col) or not area_is_clear(ship_row, ship_col):
                return False
        return True

    def place_ship(size):
        placed = False
        attempts = 0
        # Safety break for placement logic
        max_attempts = 1000 
        while not placed and attempts < max_attempts:
            start_row = random.randint(0, GRID_SIZE - 1)
            start_col = random.randint(0, GRID_SIZE - 1)
            direction = random.choice(["H", "V"])
            attempts += 1

            # Check to make sure that the ship fits in the grid
            if direction == "H" and start_col + size > GRID_SIZE:
                continue
            if direction == "V" and start_row + size > GRID_SIZE:
                continue

            if can_place(start_row, start_col, size, direction):
                for i in range(size):
                    ship_row = start_row + (i if direction == "V" else 0)
                    ship_col = start_col + (i if direction == "H" else 0)
                    board[ship_row, ship_col] = 1
                placed = True

        if not placed:
            # If placement fails after many attempts, reset the board and try again (recursive call)
            raise RuntimeError("Could not place all ships. Check placement logic.")

    for size in SHIP_SIZES:
        place_ship(size)

    return board

# setup the initial probability grid
def init_grid():
    # Initialize all cells to uniform probability 1.0
    return np.ones((GRID_SIZE, GRID_SIZE), dtype=float)

#update the probabilities based on hits and misses
def update_probabilities(grid, hits, misses):
    # Reset grid probabilities
    grid[:, :] = 1.0 

    # Set probability of missed cells to 0
    for miss_row, miss_col in misses:
        grid[miss_row, miss_col] = 0.0
    
    # Set probability of hit cells to 0, as they cannot be hit again
    for hit_row, hit_col in hits:
        grid[hit_row, hit_col] = 0.0

    # Increase probability around successful hits (Hunt mode)
    for hit_row, hit_col in hits:
        # Check adjacent cells (North, South, East, West)
        for d_row, d_col in [(-1,0),(1,0),(0,-1),(0,1)]:
            new_row = hit_row + d_row
            new_col = hit_col + d_col

            # check to see if the new row and column are within bounds and not already hit or missed
            if 0 <= new_row < GRID_SIZE and 0 <= new_col < GRID_SIZE:
                cell = (new_row, new_col)
                if cell not in hits and cell not in misses:
                    # Give adjacent cells a high probability boost
                    grid[new_row, new_col] += 4.0

    # Normalize the grid so the sum of all probabilities is 1
    total = np.sum(grid)
    if total > 0:
        grid /= total

    return grid

# choose the next shot based on the updated probabilities
def choose_next_shot(grid, hits, misses):
    flat = grid.flatten()
    used_cells = set(hits + misses)

    weights = []
    for index in range(GRID_SIZE * GRID_SIZE):
        row, col = divmod(index, GRID_SIZE)
        # Weight is 0 if the cell has already been used, otherwise it's the calculated probability
        weights.append(0 if (row, col) in used_cells else flat[index])

    # Randomly choose a cell based on the calculated weights
    chosen_index = random.choices(range(GRID_SIZE * GRID_SIZE), weights=weights, k=1)[0]
    return divmod(chosen_index, GRID_SIZE)

# convert (row, col) to Battleship notation (e.g., (1, 4) -> B5)
def coord_to_battleship(cell):
    row, col = cell
    letter = chr(ord('A') + row)
    # Columns are 1-indexed in Battleship (0 -> 1)
    number = col + 1
    return f"{letter}{number}"

# show the grid with every move until all ships are sunk
def show_grid(grid, hits, misses, next_shot, board, turn):
    # Colors: R, G, B values from 0.0 to 1.0
    
    # Deep water blue (Background)
    color_grid = np.full((GRID_SIZE, GRID_SIZE, 3), [0.2, 0.4, 1.0]) 

    for row in range(GRID_SIZE):
        for col in range(GRID_SIZE):
            # Display remaining ship segments in green
            if board[row, col] == 1 and (row, col) not in hits:
                # green for ship parts
                color_grid[row, col] = [0.0, 0.7, 0.0] 

    # Red hits
    for hit_row, hit_col in hits:
        color_grid[hit_row, hit_col] = [1.0, 0.0, 0.0] 
    # Light gray misses
    for miss_row, miss_col in misses:
        color_grid[miss_row, miss_col] = [0.8, 0.8, 0.8]

    # Highlight the next chosen shot in MAGENTA
    if next_shot and next_shot not in hits and next_shot not in misses:
        shot_row, shot_col = next_shot
        # Bright Magenta color for the next shot
        color_grid[shot_row, shot_col] = [1.0, 0.0, 1.0] 

    fig, ax = plt.subplots(figsize=(6,6))
    
    # Use imshow, which places Row 0 (A) at the top by default
    ax.imshow(color_grid)
    
    # Set X/Y limits and aspect ratio
    ax.set_xlim(-0.5, GRID_SIZE - 0.5) 
    ax.set_ylim(GRID_SIZE - 0.5, -0.5) 
    # Ensure the cells are perfect squares
    ax.set_aspect('equal') 
     
    # Define the positions for the grid lines
    line_positions = np.arange(-0.5, GRID_SIZE + 0.5, 1)

    # Draw horizontal lines (rows) and vertical lines (columns)
    ax.hlines(line_positions, -0.5, GRID_SIZE - 0.5, color='black', linewidth=1.0)
    ax.vlines(line_positions, -0.5, GRID_SIZE - 0.5, color='black', linewidth=1.0)

    # Disable default grid lines and ticks to prevent conflict
    ax.tick_params(which='both', length=0)
    ax.xaxis.grid(False)
    ax.yaxis.grid(False)
 
    # X-Labels (1 to 10 for Battleship notation)

    ax.set_xticks(np.arange(GRID_SIZE), minor=False)

    ax.set_xticklabels([str(i + 1) for i in range(GRID_SIZE)]) 
    ax.xaxis.set_label_position('top')
    ax.xaxis.tick_top()
    
    # Y-Labels (A to J)

    ax.set_yticks(np.arange(GRID_SIZE), minor=False)
    letters = [chr(ord('A') + i) for i in range(GRID_SIZE)]

    ax.set_yticklabels(letters, va='center') 

    plt.title(f"Battleship Game State (Turn {turn})")
    plt.show()

# multiple runs of the game to gather statistics
def run_game(visualize=False):
    """Runs a single simulation of the Battleship AI game."""
    try:
        board = place_ships()
    except RuntimeError:
        # If ship placement fails, return a high number of turns to indicate failure
        return 1000
        
    hits = []
    misses = []

    # Total number of ship segments to sink
    total_ship_cells = np.sum(board)
    sunk_cells = 0
    turn = 1

    while sunk_cells < total_ship_cells:
        grid = init_grid()
        grid = update_probabilities(grid, hits, misses)

        next_shot = choose_next_shot(grid, hits, misses)
        shot_row, shot_col = next_shot

        if visualize:
            print(f"Turn {turn}: Chosen shot {coord_to_battleship(next_shot)}", end=" ")
            # Show the board state *before* the shot is resolved
            show_grid(grid, hits, misses, next_shot, board, turn)
        
        # Resolve the shot
        if board[shot_row, shot_col] == 1:
            if visualize:
                print("HIT!")
            hits.append((shot_row, shot_col))
            sunk_cells += 1
        else:
            if visualize:
                print("MISS.")
            misses.append((shot_row, shot_col))

        turn += 1

    # Return the total number of turns taken to sink all ships
    return turn - 1

#main loop of project
if __name__ == "__main__":
    NUM_SIMULATIONS = 1000
    turn_counts = []

    print(f"--- Running {NUM_SIMULATIONS} Simulations for Statistical Analysis ---")
    
    # Run all simulations without visualization to get raw data
    for i in range(NUM_SIMULATIONS):
        result = run_game(visualize=False)
        turn_counts.append(result)
        # Simple progress tracking
        if (i + 1) % 10 == 0:
            print(f"Completed {i + 1}/{NUM_SIMULATIONS} runs...")

    turn_counts_np = np.array(turn_counts)
    
    # Calculate statistics
    average_turns = np.mean(turn_counts_np)
    min_turns = np.min(turn_counts_np)
    max_turns = np.max(turn_counts_np)
    
    print("\n--- Simulation Results ---")
    print(f"Total Ship Segments (Optimal Minimum Shots): {sum(SHIP_SIZES)}")
    print(f"Average Turns to Win: {average_turns:.2f} (Shots)")
    print(f"Best Run (Min Turns): {min_turns}")
    print(f"Worst Run (Max Turns): {max_turns}")
    
    # Plotting the distribution of turn counts
    plt.figure(figsize=(8, 5))
    plt.hist(turn_counts_np, bins=range(min_turns, max_turns + 2), edgecolor='black', color='teal')
    plt.title(f'Distribution of Turns to Sink All Ships ({NUM_SIMULATIONS} Runs)')
    plt.xlabel('Number of Turns (Shots)')
    plt.ylabel('Frequency')
    plt.axvline(average_turns, color='red', linestyle='dashed', linewidth=2, label=f'Average: {average_turns:.2f}')
    plt.legend()
    plt.grid(axis='y', alpha=0.5)
    plt.show()
    
    print("\n--- Running Single Visualized Game ---")
    # Run one final game with visualization so you can still step through it
    final_turns = run_game(visualize=True)
    print(f"\nSingle Visualized Game Result: All ships sunk in {final_turns} turns!")
```

# Relation to Discrete Math


This project applies several concepts from the course including using logical inference. If the move is a miss, then it cannot contain a ship which can be expressed with truth tables. The game states of hit = 1, miss = 0 can be represented as boolean variables and operations such as NOT, and OR are used to update the grid. The entire battleship is on a grid, and it uses adjacent reasoning where if a hit occurs the AI increases probability that any of the cells around are a hit. The AI decisions are using recursion. After each move it reevaluates the grid and updates probabilities until all the ships are found.
